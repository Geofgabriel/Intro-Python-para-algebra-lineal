# Figuras en Python

Sin lugar a dudas algo que nos va a interesar es poder visuzalizar nuestros resultados o los datos que tenemos de entrada. Para esto Python también ofrece una gran variedad de opciones. Siguiendo con Sympy, podemos realizar gráficos de nuestras soluciones de la siguiente manera:

```python
from sympy.plotting import plot 
from sympy import * 
x=Symbol('x') 

func = x**2

plot(func, (x,-20,20),line_color='red')
```

Ahora supongamos que queremos graficar muchas funciones en el mismo gráfico.

```python
func = x**2
func2 = (x+2)**2

p = plot(func, (x,-20,20),line_color='red',show=False)
p1 = plot(func2, (x,-20,20),line_color='blue',show=False)
p.extend(p1)
p.show()
```

y con mas funciones se procede igual...



```python
a=2; b = 1

func = exp(a*x)
func2 = exp(b*x)
func3 = func+func2
p = plot(func, (x,0,3),line_color='red',show=False)
p1 = plot(func2, (x,0,3),line_color='blue',show=False)
p2 = plot(func3, (x,0,3),line_color='green',show=False)
p.extend(p1)
p.extend(p2)
p.show()
```

Otro ejemplo...

```python
a=1; b = 1

func = cos(a*x)
func2 = sin(b*x)
func3 = func+func2

p = plot(func, (x,0,10),line_color='red',show=False)
p1 = plot(func2, (x,0,10),line_color='blue',show=False)
p2 = plot(func3, (x,0,10),line_color='green',show=False)
p.extend(p1)
p.extend(p2)
p.show()
```
